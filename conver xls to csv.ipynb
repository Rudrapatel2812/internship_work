```python
import os
import glob
import pandas as pd

input_folder = '/Users/patelrudra/Documents/Medkart/wms/credit note/Credit Note'
output_folder = '/Users/patelrudra/Documents/Medkart/wms/credit note/Credit Notes-csv'

# Get all XLS files in the input folder
xls_files = glob.glob(os.path.join(input_folder, '*.xls'))

for xls_file in xls_files:
    # Load the XLS file into a pandas DataFrame
    df = pd.read_excel(xls_file,skiprows=8)
    
    # Extract the filename without extension
    file_name = os.path.splitext(os.path.basename(xls_file))[0]
    
    # Generate the CSV file path
    csv_file = os.path.join(output_folder, file_name + '.csv')
    
    # Save the DataFrame as CSV
    df.to_csv(csv_file, index=False)

    print(f"Converted {xls_file} to {csv_file}")

print("Conversion completed.")
```
